### importing Libraries 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Importing Dataset

```python
df = pd.read_csv("data/Salary.csv")
df.head()
```

### Data Analysis

```python
#Checking the basic parameters of the dataset
df.describe()
```

```python
#checking the data types of the columns
df.dtypes
```

```python
#checking dataframe information
df.info()
```

```python
#Checking for Unique, Missing Values, and Data Types of each column
for i in df.columns:
    print(f"Column: {i}")
    print(f"Unique Values: {df[i].nunique()}")
    print(f"Missing Values: {df[i].isnull().sum()}")
    print(f"Data Type: {df[i].dtype}")
    print("-"*50)
```

```python
#Visualising salary distribution with box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Salary', data=df)
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.show()
```

```python
#Visualising years of experience with box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Years of Experience', data=df)
plt.title('Years of Experience Distribution')
plt.xlabel('Years of Experience')
plt.show()
```

Observation: Found Outliers on Years of Experience

```python
#Visualising years of experience vs. Salary with box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Years of Experience', y='Salary', data=df)
plt.title('Years of Experience vs. Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

Observation:
1. There are multiple outliers by years of experience
2. Median Salary above 17 years of experience is slightly decreasing and becoming nearly constant after 23 years of experience
3. Min salary for 31 year of experience is near about min salary of 15 years experience employee

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Years of Experience', y='Salary', hue='Gender',data=df)
plt.title('Years of Experience vs. Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

Observation:
1. Relationship between years of exp. and salary for male and female is near to lenier till 16-17 years of experience
2. Salary of Female employees are more fluctuating than male.
3. Salary for experience from 17 years onwards is near to constant

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Age', y='Salary', hue='Gender', data=df)
plt.title('Age vs. Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

Observation:
1. Relationship between age and salary is nearly linear for male and female both gender.

```python
plt.Figure(figsize=(10,6))
plt.title('Education Level vs Salary')
plt.xlabel('Education Level')
plt.ylabel('Salary')
sns.barplot(x='Education Level', y='Salary', hue='Gender', data=df)
plt.show()
```

Observation:
1. Higher the Education Level higher the salary
2. Sudden jump in the salary is observed for Education Level 0 to 1
3. Education Level 2 have more gap in salary between male and female

```python
plt.Figure(figsize=(10,6))
plt.title('Country vs Salary')
plt.xlabel('Country')
plt.ylabel('Salary')
sns.barplot(x='Country', y='Salary', hue='Gender', data=df)
plt.show()
```

Observation:
1. Salary by Country is nearly the same for all country
2. Salary in Canada for Male is slightly higher than the others
3. Salary in China for Female is slightly higher than the others

```python
plt.figure(figsize=(15,6))
plt.title('Race vs Salary')
plt.xlabel('Race')
plt.ylabel('Salary')
sns.barplot(x='Race', y='Salary', hue='Gender', data=df)
plt.show()
```

Observation:
1. Salary by Race is almost the same

```python
plt.figure(figsize=(15,6))
plt.title('Senior vs Salary')
plt.xlabel('Senior')
plt.ylabel('Salary')
sns.barplot(x='Senior', y='Salary', hue='Gender', data=df)
plt.show()
```

Observation:
1. Senior Persons have more salary than the juniors
